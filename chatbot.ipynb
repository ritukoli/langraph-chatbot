```python
from dotenv import load_dotenv
from typing import TypedDict , Annotated
from langgraph.graph import StateGraph , START, END
from langchain.chat_models import init_chat_model
from langgraph.graph.message import add_messages
load_dotenv()
```

```python
llm = init_chat_model ("google_genai:gemini-2.5-flash")

class chatbot (TypedDict) :
    message : Annotated[list , add_messages]


def chatbot1(state : chatbot) -> chatbot :
    return{"message" :[llm.invoke(state["message"])]} 

builder = StateGraph(chatbot)

builder.add_node("chatbot_node" , chatbot1)
builder.add_edge(START, "chatbot_node")
builder.add_edge("chatbot_node", END)

graph = builder.compile()
```

```python
message = {"role" : "user" , "content" : "what color is an apple? just give a name"}

response= graph.invoke({"message" : [message]})

response['message']
```

```python
state = None
while True :
    input_message = input("you: ")
    if input_message.lower() in ("quit", "exit") :
        break
    if state is None :
        state : chatbot={
            "message" : [{"role" : "user" , "content" : input_message}]
        }
    else:
        state["message"].append({"role" : "user" , "content" : input_message})

    state = graph.invoke(state)
    print("bot: " , state["message"][-1].content)
```
